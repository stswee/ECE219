```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/radraid2/dongwoolee/project4/subsample_preprocessed_tweets.csv")
```

```python
times = df['unix_time']
df["tweet_text_length"] = df["tweet_text"].str.len()

tweets_per_hour = np.zeros(int((max(times)-min(times))/3600+1))
start = min(times)
for i in times:
    tweets_per_hour[int((i-start)/3600)] += 1

# Use barplot
# Get each hour
plt.figure(figsize=(4,3))
x = np.linspace(0, len(tweets_per_hour)-1, len(tweets_per_hour))
plt.bar(x, tweets_per_hour)
plt.title("Number of Tweets throughout the Day of Superbowl")
plt.xlabel("Hour")
plt.ylabel("Number of tweets")
plt.show()

plt.figure(figsize=(4, 3))
sns.histplot(df["retweet_count"], bins=20, kde=True, log_scale=(True, False))
plt.title("Distribution of Retweet Count (Log Scale)")
plt.xlabel("Retweet Count (log scale)")
plt.ylabel("Frequency")
plt.ylim(0, 2e6)  # Adjust based on your data
plt.show()


plt.figure(figsize=(4, 3))
sns.histplot(df["favorite_count"] + 1, bins=20, kde=True, log_scale=(True, False))
plt.title("Distribution of Favorite Count (Log Scale)")
plt.xlabel("Retweet Count (log scale)")
plt.ylabel("Frequency")
plt.ylim(0, 2e6)  # Adjust based on your data
plt.show()

# Plot the distribution of tweet text lengths
plt.figure(figsize=(4, 3))
sns.histplot(df["tweet_text_length"], bins=20, kde=True)
plt.title("Tweet Text Length Distribution")
plt.xlabel("Tweet Text Length (characters)")
plt.ylabel("Frequency")
plt.show()
```
